```python
from google.colab import userdata
api_key=userdata.get('BING_API_KEY')
```

```python
import requests

def bing_search(query, api_key, max_results=5):
    url = "https://google.serper.dev/search"
    headers = {
        "X-API-KEY": api_key,
        "Content-Type": "application/json"
    }
    payload = {
        "q": query
    }
    response = requests.post(url, headers=headers, json=payload)
    data = response.json()
    results = data.get("organic", [])[:max_results]
    return results

# Ví dụ:
results = bing_search("Thủ tướng Việt Nam tham dự hội nghị G7 năm 2024", api_key)

```

```python
import time
def calculate_time(function):
  start_time=time.time()
  result=function()
  end_time=time.time()
  total_time=end_time-start_time
  print(f"Time to run function{function.__name__}:{total_time}")
  return (result,total_time)
```

```python
def show_result(result):
  for r in result:
    print(f"- {r['title']}\n  {r['snippet']}\n  {r['link']}\n")
```

```python
list_results=[]
list_query = [
    "Thủ tướng Việt Nam tham dự hội nghị G7 năm 2024",
    "Giá xăng dầu ở Việt Nam tăng mạnh trong tháng 5 năm 2024",
    "VinFast ra mắt mẫu ô tô điện mới tại Mỹ",
    "Nắng nóng kỷ lục xảy ra tại Hà Nội trong tháng 4 năm 2024",
    "Giải vô địch bóng đá U23 châu Á 2024: Việt Nam vào bán kết"
]
list_time_search=[]
def evaluate_bing_search(api_key, max_results=5):
  for query in list_query:
    print(f"Query: {query}")
    result,time_search=calculate_time(lambda: bing_search(query, api_key, max_results))
    show_result(result)
    list_results.append(result)
    list_time_search.append(time_search)
```

```python
evaluate_bing_search(api_key)
```

```python
import pandas as pd
# Tạo DataFrame
df = pd.DataFrame({
    'query': list_query,
    "Hit@5":["100%","100%","100%","100%","100%"],
    "Precision@1":["100%","100%","100%","100%","100%"],
    'time_search': list_time_search,
    'results': list_results
})
print("N=5")
df.head()
```

```python
average_time_search = df['time_search'].mean()
print(f"\nAverage time for search queries: {average_time_search} seconds")
```
